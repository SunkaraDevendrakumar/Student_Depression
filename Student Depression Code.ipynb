```python
df.drop(columns="id",inplace=True)
```

```python
df['City'] = df['City'].replace([
    "'Less Delhi'", "'Less than 5 Kalyan'", 'Saanvi', 'Mira', 'Harsha', 'Vaanya', 
    'Gaurav', 'Harsh', 'Reyansh', 'Kibara', 'Rashi', 'Mihir', 'Nalini', 'Nandini',
    'Khaziabad', 'Nalyan','M.Tech', 'Bhavna', '3.0', 'ME', 'M.Com'], 'Unknown')
df['City'] = df['City'].replace('City', 'Unknown')

df['Profession'] = df['Profession'].replace([
    "'Civil Engineer'", "'UX/UI Designer'", "'Digital Marketer'", "'Content Writer'", 
    "'Educational Consultant'", "'Manager'", "'Chef'", "'Doctor'", "'Lawyer'", 
    "'Entrepreneur'", "'Pharmacist'"], 'Unknown') 
```

```python
df['Sleep Duration'] = df['Sleep Duration'].replace({
    "'5-6 hours'": '5-6 hours', "'Less than 5 hours'": 'Less than 5 hours', 
    "'7-8 hours'": '7-8 hours', "'More than 8 hours'": 'More than 8 hours'
})
```

```python
df['Financial Stress'] = df['Financial Stress'].replace('?', 'NaN')  # Or replace with a numerical value like the mean or mode
df['Financial Stress'] = pd.to_numeric(df['Financial Stress'], errors='coerce')  # Convert to numeric, handle errors as NaN
```

```python
df['Degree'] = df['Degree'].replace([
    "'Class 12'", "'B.Sc'", "'M.Tech'", "'MBA'", "'Ph.D'", 'ME', 'Others'], 'Unknown')  # Handle noise
```

```python
df['Have you ever had suicidal thoughts ?'] = df['Have you ever had suicidal thoughts ?'].replace({
    "'Yes'": 'Yes', "'No'": 'No'
})
```

```python
# Drop rows with 'Unknown' in any important columns (optional)
df = df[df['City'] != 'Unknown']
df = df[df['Profession'] != 'Unknown']

# Or, replace missing values with a default
df.fillna(df.mode().iloc[0], inplace=True)  # Fill with the mode (most common value)
```

```python
df['Dietary Habits'] = df['Dietary Habits'].replace('Others', np.nan)
df['Sleep Duration'] = df['Sleep Duration'].replace('Others', np.nan)
df['Dietary Habits'].fillna(df['Dietary Habits'].mode()[0], inplace=True)
df['Sleep Duration'].fillna(df['Sleep Duration'].mode()[0], inplace=True)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Depression', axis=1)
y = df['Depression']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test) 
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.95)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)  
```

```python
X_test_scaled = scaler.transform(X_test)    
X_test_pca = pca.transform(X_test_scaled)  
```

```python
import matplotlib.pyplot as plt

# Apply PCA to the original training data
pca = PCA(n_components=2)
X_orig_pca = pca.fit_transform(X_train)

plt.figure(figsize=(7, 5))
plt.scatter(X_orig_pca[:, 0], X_orig_pca[:, 1], c=y_train, cmap='coolwarm', alpha=0.6)
plt.title(" Before SMOTE - PCA")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.show()
```

```python
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_pca, y_train)
```

```python
y_train_smote.value_counts(normalize=True)
```

```python
from collections import Counter
print("Before SMOTE:", Counter(y_train))
print("After SMOTE:", Counter(y_train_smote))
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train_smote)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_train_smote, cmap='coolwarm', alpha=0.6)
plt.title("SMOTE - After Oversampling")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
Rf_model=RandomForestClassifier()
Rf_model.fit(X_train_smote,y_train_smote)
```

```python
y_predrf=Rf_model.predict(X_test_pca)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
conf_matrix = confusion_matrix(y_test, y_predrf)
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
print(classification_report(y_test,y_predrf))
```

```python
from sklearn.model_selection import cross_val_score

cross_val_scores = cross_val_score(Rf_model, X_train_smote, y_train_smote, cv=5)
print(f"Cross-validation scores: {cross_val_scores}")
print(f"Mean CV score: {cross_val_scores.mean()}")
```
